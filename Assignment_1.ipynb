##Q2 : Write a Python program to find the largest of three numbers.
The program should take three integer inputs from the user and print the largest number

```python
def find_largest(num1, num2, num3):
    if (num1 >= num2 and num1 >= num3):
        return num1
    elif (num2 >= num1 and num2 >= num3):
        return num2
    else:
        return num3

a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
c = int(input("Enter the third number: "))


largest = find_largest(a, b, c)


print(f"The largest number is {largest}")
```

## Q1 : Create a NumPy array of shape (3, 3) filled with random numbers.
  Then, compute the mean, median, and standard deviation of the array.

```python
import numpy as np

array = np.random.rand(3,3)

mean_value = np.mean(array)

median_value = np.median(array)

std_deviation = np.std(array)

print("Array : \n",array)
print(f"Mean : {mean_value}")
print(f"Median : {median_value}")
print(f"Standard Deviation : {std_deviation}")
```

## **4.Matplotlib**

## Q1 : Create a line plot using Matplotlib.
Plot the function y = x^2 for x values ranging from -10 to 10. Label the axes and give the plot a title.

```python
import matplotlib.pyplot as plt

x = np.arange(-10,11,1)
##x

y = x**2
##y


plt.plot(x,y,label = 'y=x^2')

plt.xlabel ('x')
plt.ylabel ('y')

plt.title('y = x^2 plot')



plt.grid(True)

plt.show()
```

## Q 2: Create a heatmap using Seaborn.
Use a correlation matrix of a dataset to create a heatmap. Annotate the heatmap for better readability

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df = pd.read_csv('/content/insurance.csv')

df.sex = le.fit_transform(df.sex)

df.smoker = le.fit_transform(df.smoker)

df1 = pd.get_dummies(df,columns=['region'])


correlation = df1.corr()


sb.heatmap(correlation, annot=True, cmap="coolwarm")


plt.show()
```
